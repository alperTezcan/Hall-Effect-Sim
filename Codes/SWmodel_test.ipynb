```python
plt.figure(figsize=(24, 8))

for ang in [5, 30, 60, 90, 120, 150, 178, 179, 179.5, 180, 180.5, 181]:
    M = model.calculate_hysteresis(ang, 0.1)

    plt.subplot(131)
    plt.plot(M[:, 0], M[:, 1], linestyle="-", label=f'{ang}')
    plt.xlabel("H")
    plt.ylabel("Mx")
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)

    plt.subplot(132)
    plt.plot(M[:, 0], M[:, 2], linestyle="-", label=f'{ang}')
    plt.xlabel("H")
    plt.ylabel("My")
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)        

    plt.subplot(133)
    plt.plot(M[:, 0], M[:, 3], linestyle="-", label=f'{ang}')
    plt.xlabel("H")
    plt.ylabel("Mz")
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)

plt.legend()
plt.suptitle('Magnetization Components')
plt.show()
```
